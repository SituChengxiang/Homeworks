```python
import numpy as np

A = np.array([[-5, 2, -1],
              [ 1, 0,  3],
              [ 3, 1,  6]], dtype=float)

try:
    from scipy.linalg import lu
    P, L, U = lu(A)
    print("Using scipy.linalg.lu:")
    print("P:\n", P)
    print("L:\n", L)
    print("U:\n", U)
except Exception:
    # Doolittle without pivoting (works since A[0,0] != 0)
    n = A.shape[0]
    L = np.eye(n)
    U = np.zeros((n, n))
    for i in range(n):
        for k in range(i, n):
            U[i, k] = A[i, k] - sum(L[i, j] * U[j, k] for j in range(i))
        for k in range(i + 1, n):
            L[k, i] = (A[k, i] - sum(L[k, j] * U[j, i] for j in range(i))) / U[i, i]
    P = np.eye(n)
    print("Using custom Doolittle (no pivot):")
    print("P:\n", P)
    print("L:\n", L)
    print("U:\n", U)

# 验证 PA = LU
print("\nCheck P @ A:\n", P @ A)
print("Check L @ U:\n", L @ U)
```
